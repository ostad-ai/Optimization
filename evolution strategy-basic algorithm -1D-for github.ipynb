```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
# 1D function that we seek its minimum
def f(x):
    return x**2+2*x-4.
```

```python
# basic evolution strategy: 1D
def evolution_strategy_1d(func,x0=0.,
                        maxIter=100):
    x=x0
    for i in range(maxIter):
        y=x+np.random.normal()
        if func(y)<func(x): x=y
    return x,func(x)
```

```python
x0=10. #initial guess
# find minimum point x* for f(x) 
x_star,y_star=evolution_strategy_1d(f,x0)
xs=np.linspace(-10,10,100)
plt.plot(xs,f(xs),'-')
plt.plot(x_star,y_star,marker='$\U0000263C$',ms=10,
         c='red',label='minimum')
plt.title('Evolution strategy for 1D minimization')
plt.xlabel('$x$',fontsize=15)
plt.ylabel('$f(x)$',fontsize=15)
plt.legend(); plt.show()
```
