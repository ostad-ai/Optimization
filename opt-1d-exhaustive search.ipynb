## Optimization
### Exhaustive Search, a one-dimensional optimization algorithm 
We find the minimum of a **unimodal function** within a given interval $[a,b]$.
 - In this method, the given interval $[a,b]$ is used to get *n* equally spaced points $x_i$ including $a$ and $b$, in an increasing order.
 - Then, the minimum point, say $x_m$, for which the function is minimum among the $n$ points is obtained. 
 - Next, the new interval is formed by points that are left and right to the current minimum. 
 - The whole process can be repeated until we are satisfied with the result.
 
**Hint:** A *unimodal function* is the function with just one minimum or just one maximum. Here, we are interested in unimodal functions having a single minimum. However, we can have unimodal functions having a single maximum.

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
# the one-dimensional function to minimize
def f1(x):
    return x**2-5*x+3
```

```python
# give the function to minimize: func
# and the initial interval to search into: [a, b]
# and the number of points in each iteration: n
# finally, the number of iterations: iter
def exhaustive_search(func,a=0,b=1,n=10,iter=25):
    xa,xb=a,b
    for _ in range(iter):
        # returns n equally-spaced points from interval
        xs=np.linspace(xa,xb,n) 
        arg_min=np.argmin(func(xs))
        arg_min_a=arg_min-1
        arg_min_b=arg_min+1
        if arg_min_a<0:
            arg_min_a=0
        if arg_min_b>=n:
            arg_min_b=n-1
        xa,xb=xs[arg_min_a],xs[arg_min_b]
    xmin=xs[arg_min]
    return xmin,func(xmin),xs[arg_min_a],xs[arg_min_b]
```

```python
# example: to minimize f1 
a,b=0,10 #interval [a, b]
xmin_exact=2.5
xmin,f_min,a_final,b_final=exhaustive_search(f1,a,b)
print('----The minimum found by Exhaustive search:')
print(f'The x*={xmin} yields function(x*)={f_min}')
print(f'Hint: We know the exact x* is: {xmin_exact}')
```

```python
# display the function f1 and its minimum found by exhaustive search
xs=np.linspace(a,b,100)
plt.plot(xs,f1(xs),'-',label=f'x*={xmin}\n$f_1(x*)=${f_min}')
plt.plot(xmin,f_min,'d')
plt.title(f'The graph of $f_1(x)=x^2-5x+3$ and its minimum')
plt.legend()
plt.show()
```
