## Convex sets and convex functions
A set is **convex** if the line segment joining any two points of the set lies entriely in the set.
<br> A function is **convex** if its domain is convex and the line segment joining
any two points $(x,f(x))$ and $(y,f(y))$ lies above the graph of $f$ between the two points.
<br>**Reminder:** The first order Taylor approximation of $f$ around $\boldsymbol{x}$ denoted by $L$ is:
<br> $L(\boldsymbol{x},\boldsymbol{y})=f(\boldsymbol{x})+\nabla f(\boldsymbol{x})^T(\boldsymbol{y}-\boldsymbol{x})$
<br>**Hint 1:** For a convex function, its first-order Taylor approximation always gives an under-estimate of the actual value of $f$ at the point of interest, $\boldsymbol{x}$. Specifically:
<br>$f(\boldsymbol{y})\ge L(\boldsymbol{x},\boldsymbol{y})$
<br> **Hint 2:** If $\nabla f(x^*)=0$, then from the above inequality we infer that $x^*$ is the global minimizer of $f$.
<br>**Hint 3:** A twice differentiable function $f$ is convex if and only if its domain is convex and its Hessian, $\nabla^2f$ is positive semidefinite.
<br> The above property for function $f:R \rightarrow R$ reduces to $f''(x)\ge0$ for all $x$ in $dom f$.
<br> **Hint 4:** We say $f$ is **concave** if $-f$ is convex.
<br> **Hint 5:** A function can be neither convex nor concave.

```python
# importing required modules
import numpy as np
from matplotlib import pyplot as plt
```

For function $f:R \rightarrow R$, the first order Taylor approximation becomes:
<br>$f(x)=f(x_0)+f'(x_0)(x-x_0)$

Thus, the linear approximator for $f(x)=x^2$ at point $x_0$
<br>$\nabla f(x_0)=f'(x_0)=2x_0$
<br>$f(x)=f(x_0)+2x_0(x-x_0)$
<br>$y=x^2_0+2xx_0-2x^2_0$
<br>$y=2xx_0-x^2_0$

```python
# Example: displaying two global under-estimators 
# of the convex function f(x)=x^2
for x0 in [-10,7]:
    y0=x0**2
    xs=np.linspace(-15,15,100)
    ys=xs**2
    xt=np.linspace(-10,10,100)
    yt=2*xt*x0-x0**2
    plt.plot(x0,y0,'*')
    plt.plot(xs,ys,'-')
    plt.plot(xt,yt,'-')
    plt.text(x0+1,y0,f'({x0}, {y0})')
plt.title('A convex function & its two'+\
' global under-estimators at the given points')
plt.show()
```

Some examples of convex functions:
<br>**Powers of absolute values**, $|x|^p$ where $p\ge1$ is convex on $R$.
<br>**Exponential**, $e^{ax}$ is convex for any $a\in R$.
<br> **Powers**, $x^a$ for $x\in R_{++}$ is convex when $a\ge1$ or $a\le0$. $R_{++}$ denotes positive real numbers.
<br> **Norms**, every  norm on $R^n$ is convex.
<br> **Max functions**, $f(x)=max\{x_1,x_2,...,x_n\}$ is convex on $R^n$.
<br> **Linear functions**, $f(x)=ax+b$ is both convex and concave.

```python
# Example: we display |x|^p for p>=1
xs=np.linspace(-20,20)
for p in [1.,1.1,1.5,1.7,2.]:
    ys=np.abs(xs)**p
    plt.plot(xs,ys,'-',label='y=$|x|^{%s}$'%(str(p)))
plt.title(f'Convex functions $|x|^p$, where $p>=1$')
plt.legend(); plt.show()
```

```python
# Example: we display |x|^p for p<1
xs=np.linspace(-5,5,10000)
for p in [-.1,.05,.1,.2,.5]:
    ys=np.abs(xs)**p
    plt.plot(xs,ys,'-',label='y=$|x|^{%s}$'%(str(p)))
plt.title(f'Concave functions $|x|^p$, where $p<1$')
plt.legend();plt.show()
```
